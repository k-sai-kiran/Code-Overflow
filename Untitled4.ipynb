```python
import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def isbright(image, dim=10, thresh=0.5):
    
    image = cv2.resize(image, (dim, dim))
    
    L, A, B = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2LAB))
  
    L = L/np.max(L)

    return np.mean(L) > thresh
```

```python
image = cv2.imread("img/IMG_20210624_200022.jpg")

# find if image is bright or dark
text = "bright" if isbright(image) else "dark"

# write text on image
cv2.putText(image, "{}".format(text), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

# show image
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
